## 立体匹配

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt


if __name__ == "__main__":
    window_size =5
    min_disp = 16
    num_disp =192-min_disp
    blockSize = window_size
    uniquenessRatio = 1
    speckleRange = 3
    speckleWindowSize = 3
    disp12MaxDiff = 200
    P1 = 600
    P2 = 2400
    imgL = cv2.imread("./image/imgL.jpg")
    imgR = cv2.imread("./image/imgR.jpg")
    # imgL = cv2.resize(imgL, (0, 0), fx=0.5, fy=0.5)
    # imgR = cv2.resize(imgR, (0, 0), fx=0.5, fy=0.5)

    # 创建一个StereoSGBM实例
    stereo = cv2.StereoSGBM_create(
        minDisparity = min_disp,
        numDisparities=num_disp,
        blockSize = window_size,
        uniquenessRatio = uniquenessRatio,
        speckleRange = speckleRange,
        speckleWindowSize = speckleWindowSize,
        disp12MaxDiff = disp12MaxDiff,
        P1 = P1,
        P2 = P2
    )

    print('computing disparity')
    disp = stereo.compute(imgL,imgR).astype(np.float32)/16.0
    # 中值滤波去除视差图中的噪声
    # disp = cv2.medianBlur(disp.astype(np.uint8), 25)

    disp_f = (disp-min_disp)/num_disp


    cv2.namedWindow('disparity', cv2.WINDOW_FREERATIO)
    cv2.imshow('disparity',disp_f)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    _, axs = plt.subplots(1, 3, figsize=(15, 5))
    axs[0].imshow(imgL)
    axs[0].set_title('L')
    axs[0].axis('off')
    axs[1].imshow(imgR)
    axs[1].set_title('R')
    axs[1].axis('off')
    axs[2].imshow(disp_f)
    axs[2].set_title('Result')
    axs[2].axis('off')
    # axs[3].imshow(disp_f)
    # axs[3].set_title('Result2')
    # axs[3].axis('off')
    plt.show()
    
    


```
